```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
chromedriver = 'chromedriver.exe' 
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(3) 
```

```python
driver.get('https://nfds.go.kr/stat/general.do')
time.sleep(2)

area="""//*[@id="headingThree"]/a"""
driver.find_element_by_xpath(area).click()
time.sleep(2)

area1 ="""//*[@id="collapse3"]/ul/li[1]/div[1]/a"""
driver.find_element_by_xpath(area1).click()
time.sleep(2)

area2="""//*[@id="collapse3"]/ul/li[2]/div[2]/a"""
driver.find_element_by_xpath(area2).click()
```

```python
date=int(input("날짜를 입력하세요:(ex : 2017-04-03)"))

# 날짜 받아오는 형식 수정(권장)
i=0
while True:    
    elem = driver.find_element_by_name("datePicker1s")
    elem.clear()
    a=20170401
    a=a+i
    elem.send_keys(a)

    elem1 = driver.find_element_by_name("datePicker1e")
    elem1.clear()
    elem1.send_keys(a)

    search="""//*[@id="searchItem"]/div[4]/div[2]/div[2]/a[1]/button"""
    driver.find_element_by_xpath(search).click()
    time.sleep(3)
    
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    table = soup.select('table')


    table_html = str(table)
    table_df_list = pd. read_html(table_html)
    table_df = table_df_list[0]
    table_df["날짜"]= a
#     table_df=table_df.drop([0,1])
#     table_df=table_df1.drop(['구분'],axis=1)
    table_df.to_csv("file.csv",index=False, mode='a',encoding='utf-8-sig',header=False)
    i+=1
    if i == date:
        break
```
